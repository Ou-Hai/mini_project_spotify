```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/cleaned/top50_clean.csv")  

df.head()
```

```python
genre_pop = (
    df.groupby("Genre_Group")["Popularity"]
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8, 5))
genre_pop.plot(kind="bar")

plt.xlabel("Genre")
plt.ylabel("Average Popularity")
plt.title("Average Popularity by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Danceability"], df["Popularity"])

plt.xlabel("Rank")
plt.ylabel("Danceability")
plt.title("Most danceable songs by rank")

plt.tight_layout()
plt.savefig("reports/figures/top10_danceability.png", dpi=300, bbox_inches='tight')
plt.show()
```
